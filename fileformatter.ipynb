```python
### Scene name ###

scenes = {1: "Practice", 2: "Elvtest", 3: "Outside", 4: "Hallway", 5: "Elevator", 6: "Hall"}
slst = [str(x) for x in[*scenes.keys()]]
```

```python
#Folder Name Changing code
datapath = './new' #choose folder to convert

for idx, folder in enumerate(os.listdir(datapath)):
    datapath_ = os.path.join(datapath, folder)
    
    #Change Name of anxiety log file
    anxiety_filepath = glob(datapath_+ '/*.log')
    newname_anxiety = os.path.join(datapath_,'anxiety.log') 
    os.rename(anxiety_filepath[0], newname_anxiety)
    
    for in_folder in os.listdir(datapath_):
        datapath__ = os.path.join(datapath_, in_folder)

        
        if in_folder in slst:
            name = scenes[int(in_folder)]
            
            print(name)

            file_oldname_custom = os.path.join(datapath__,'CustomEvents.csv')
            file_newname_custom = os.path.join(datapath_,f'{name}_customevent.csv')
            os.rename(file_oldname_custom, file_newname_custom)
            
            file_oldname_lce = os.path.join(datapath__,'LifeCycleEvents.csv')
            file_newname_lce = os.path.join(datapath_,f'{name}_lifecycleevent.csv')
            os.rename(file_oldname_lce, file_newname_lce)
            
            file_oldname_pos = os.path.join(datapath__,'PositionData.csv')
            file_newname_pos = os.path.join(datapath_,f'{name}_position.csv')
            os.rename(file_oldname_pos, file_newname_pos)
            
            file_oldname_rot = os.path.join(datapath__,'RotationData.csv')
            file_newname_rot = os.path.join(datapath_,f'{name}_rotation.csv')
            os.rename(file_oldname_rot, file_newname_rot)
            
            file_oldname_sumet = os.path.join(datapath__,'SubjectMetaData.csv')
            file_newname_sumet = os.path.join(datapath_,f'{name}_subjectmetadata.csv')
            os.rename(file_oldname_sumet, file_newname_sumet)
            
            file_oldname_sub = os.path.join(datapath__,'Subjects.csv')
            file_newname_sub = os.path.join(datapath_,f'{name}_subjects.csv')
            os.rename(file_oldname_sub, file_newname_sub)
            
            #Remove empty folder, 0
            os.rmdir(os.path.join(datapath_,in_folder))
                

    


                

        
            
    
    
            
            
    
```
